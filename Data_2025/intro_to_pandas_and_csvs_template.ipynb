```python
# use csv to open the file
import csv

with open("/content/BMW sales data (2010-2024).csv", newline="") as f:
  # get the reader object
  reader = csv.reader(f)
  print(reader)

  # first row is the header
  header = next(reader)
  # print rows
  for row in reader:
    print(row)
```

```python
# store in dictionary
with open("/content/BMW sales data (2010-2024).csv", newline="") as f:
  reader = csv.DictReader(f)
  for row in reader:
    # treat it as a dictionary
    print(row["Model"])
```

```python
# simple data analysis
total, count = 0, 0
with open("/content/BMW sales data (2010-2024).csv", newline="") as f:
  reader = csv.DictReader(f)
  for row in reader:
    # get the average price for all the sales
    total += int(row["Price_USD"])
    count += 1

print("Average Sales", total / count)
```

```python
from collections import Counter
# simple statistics
with open("/content/BMW sales data (2010-2024).csv", newline="") as f:
  reader = csv.DictReader(f)
  rows = list(reader)

prices = [int(r["Price_USD"]) for r in rows if r["Price_USD"]]
models = [r["Model"] for r in rows if r["Model"]]
print(f"All cars Prices: {prices}")
print(f"Minimum price: {min(prices)}")
print(f"Maximum price: {max(prices)}")
print(f"Most Common Cars: {Counter(models).most_common((3))}")
```

```python
# some basic visualization (wrong)
import matplotlib.pyplot as plt

colors = [r["Color"] for r in rows if r["Color"]]

plt.bar(colors, models)
plt.xlabel("Colors")
plt.ylabel("Models")
plt.show()
```

```python
# some basic visualization (right)
color_counts = Counter(r["Color"] for r in rows if r["Color"])

plt.bar(color_counts.keys(), color_counts.values())
plt.xlabel("Color")
plt.ylabel("Number of Cars")
plt.show()
```

```python
# import csv 2D
df = pd.read_csv("/content/BMW sales data (2010-2024).csv")
df.head()
```

```python
# column indexing
df["Model"]
```

```python
# iloc vs loc (rows indexing)
# iloc -> integer-location based indexing

df.iloc[0, 1] # Row 0, Column 1 (Year Column)
df.iloc[0] # Entire first row (5 series model)
```

```python
# loc -> label-based indexing
df.loc[2, "Model"] # Row with index 2, column "Model"
```

```python
# filtering data
df[df["Year"] > 2022]

# get the cars from range 2022 - 2023 and sort by price
cars = df[(df["Year"] > 2022) & (df["Year"] <= 2023)]
cars.sort_values(by="Price_USD", ascending=False)
```

```python
# can we see it?
df.plot(kind="line")
```

```python
df.groupby("Model")["Sales_Volume"].mean().plot(kind="bar", ylabel="Average Sales")   # average sales per model
```

```python
movies_dataset = pd.read_csv("/content/imdb_movies.csv")
movies_dataset.info()
```

```python
movies_dataset.head()
```

```python
# filter the dataset to the score and budget
filtered = movies_dataset[["names", "score", "budget_x"]]
filtered.head()
```

```python
# take a look at the max and min score
min_score = filtered["score"].min()
max_score = filtered["score"].max()

# get the movies
worst_movie = filtered[filtered["score"] == min_score]
best_movie = filtered[filtered["score"] == max_score]

worst_movie[["names", "score", "budget_x"]]
```

```python
# plot budget vs score
filtered.plot.scatter(x="budget_x", y="score")
plt.xlabel("Budget (USD)")
plt.ylabel("Score")
plt.title("Movie Score vs Budget")
plt.show()
```

```python
# check correlation coefficient
corr = filtered["budget_x"].corr(filtered["score"])
print("Correlation between budget and score:", corr)
```

```python
import numpy as np
x = filtered["budget_x"]
y = filtered["score"]

# fit a line
m, b = np.polyfit(x, y, 1)  # linear fit

plt.scatter(x, y, alpha=0.6)
plt.plot(x, m*x + b, color="red")  # trend line
plt.xlabel("Budget")
plt.ylabel("Score")
plt.title("Score vs Budget with Trend Line")
plt.show()
```
